```python
import pandas as pd
import numpy as np


def clean_data(path: str):
    df = pd.read_csv(path)

    df = df.loc[df['product'] == 'pink morsel']

    df['price'] = df['price'].str.replace('$', '')
    df['price'] = df['price'].astype(np.float16)
    df['sales'] = df['quantity'] * df['price']
    df = df[['sales', 'date', 'region']]
    return df

daily0 = clean_data('./data/daily_sales_data_0.csv')
daily1 = clean_data('./data/daily_sales_data_1.csv')
daily2 = clean_data('./data/daily_sales_data_2.csv')

display(daily0.head())
display(daily1.head())
display(daily2.head())
```

```python
# convert to a single file

df = pd.concat([daily0, daily1, daily2], ignore_index=True)
df.to_csv('./data/clean_data.csv')
```
